```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv(r"E:\codsoft\dataset\task3_Churn_Modelling.csv")
data
```

```python
data = data.drop(columns=["CustomerId","RowNumber","Surname"],axis=1)
```

```python
data['Geography'].value_counts()
```

This dataset is originated from US bank so we need to consider geography

```python
data.isna().sum()
```

```python
data['Gender'] = data['Gender'] .apply(
    lambda x: 1 if x == 'Female' else 0)
```

```python
data
```

```python
from sklearn.preprocessing import OrdinalEncoder,StandardScaler
ord = OrdinalEncoder()

data[['Geography']] = ord.fit_transform(data[['Geography']])
```

```python
data
```

```python
sc = StandardScaler()
X = data.drop("Exited",axis=1)
X =sc.fit_transform(X)
y = data['Exited']
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
random_state = 42
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=random_state)
```

```python
X_train
```

```python
print(np.isnan(X).sum())
print(np.isnan(y).sum())
```

```python
# Trying all boosting techniques
# 1. Adaboost(Weak learner to strong learner)
# 2. Gradient boost(One output to other decision model)
# 3. Xgboost (Updates split from information gain)
```

```python
from sklearn.metrics import classification_report,accuracy_score
from sklearn.ensemble import AdaBoostClassifier , GradientBoostingClassifier
from xgboost import XGBClassifier
```

```python
gb_model = GradientBoostingClassifier()
gb_model.fit(X_train, y_train)
gb_preds = gb_model.predict(X_test)

print("Gradient Boosting Accuracy:", accuracy_score(y_test, gb_preds))
print(classification_report(y_test, gb_preds))
```

```python
ada = AdaBoostClassifier()
ada.fit(X_train,y_train)
ada_pred = ada.predict(X_test)

print("AdaBoost Accuracy:", accuracy_score(y_test, ada_pred))
print(classification_report(y_test, ada_pred))
```

```python
xg = XGBClassifier()
xg.fit(X_train,y_train)
xg_pred = ada.predict(X_test)

print("XgBoost Accuracy:", accuracy_score(y_test, xg_pred))
print(classification_report(y_test, xg_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)

# Predict on test data
rf_preds = rf_model.predict(X_test)
print("Random Forest Accuracy:", accuracy_score(y_test, rf_preds))
print(classification_report(y_test, rf_preds))
```

```python
from sklearn.metrics import precision_score,recall_score
precision = [
    precision_score(y_test, gb_preds),
    precision_score(y_test, ada_pred),
    precision_score(y_test, xg_pred),
    precision_score(y_test, rf_preds)
]

recall = [
    recall_score(y_test, gb_preds),
    recall_score(y_test, ada_pred),
    recall_score(y_test, xg_pred),
    recall_score(y_test, rf_preds)
]
results = {
    "Model": ["Gradient Boosting", "AdaBoost", "XGBoost", "Random Forest"],
    "Accuracy": [
        accuracy_score(y_test, gb_preds),
        accuracy_score(y_test, ada_pred),
        accuracy_score(y_test, xg_pred),
        accuracy_score(y_test, rf_preds)
    ],
    "Precision":precision,
    "Recall":recall,
}

df_results = pd.DataFrame(results)
print(df_results)
```
